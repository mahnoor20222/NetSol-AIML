## 📝 Task 1 — Multi-Line Plot with Matplotlib

### Background
Line plots are used to show **trends over time**. When comparing multiple series on the same chart you add a legend so readers can distinguish them.

### 🎯 Your Task
The table below shows the **monthly revenue** (in $000s) of two products over 6 months.

| Month | Product A | Product B |
|-------|-----------|-----------|
| Jan | 15 | 10 |
| Feb | 18 | 13 |
| Mar | 14 | 16 |
| Apr | 22 | 18 |
| May | 25 | 20 |
| Jun | 28 | 24 |

Create a **single line chart** that shows **both products** with:
- Different colors and markers for each line
- A legend identifying each product
- Proper axis labels and a descriptive title

## 📝 Task 2 — Grouped Bar Chart with Matplotlib

### Background
A **grouped bar chart** places multiple bars side by side for the same category, making it easy to compare values within each group.

### 🎯 Your Task
The data below shows **exam scores** for three subjects across two classes.

| Subject | Class A | Class B |
|---------|---------|---------|
| Math | 72 | 68 |
| Science | 85 | 80 |
| English | 78 | 74 |
| History | 65 | 70 |

Create a **grouped bar chart** with:
- Two bars (Class A and Class B) side by side for each subject
- Different colors for each class
- Value labels on top of each bar
- A legend, axis labels, and a title

> 💡 **Hint:** Use `np.arange(len(subjects))` for x-positions, then offset each group with `x - 0.2` and `x + 0.2` and `width=0.4`.  
> To add labels: loop over bars using `bar.get_height()` and `plt.text()`.


````python
import numpy as np
import matplotlib.pyplot as plt

# Data
subjects = ['Math', 'Science', 'English', 'History']
class_a  = [72, 85, 78, 65]
class_b  = [68, 80, 74, 70]

#  Create x positions
x = np.arange(len(subjects))
width = 0.4  # bar width

# Plot bars
bars_a = plt.bar(x - 0.2, class_a, width=width, color='gray', label='Class A')
bars_b = plt.bar(x + 0.2, class_b, width=width, color='orange', label='Class B')

#  Add value labels above each bar
for bar in bars_a:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, str(height), ha='center', va='bottom')

for bar in bars_b:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, str(height), ha='center', va='bottom')

#  Add labels, title, legend
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.title('Exam Scores Comparison (Class A vs Class B)')
plt.xticks(x, subjects)
plt.legend()

#  Show the plot
plt.show()
````

Histograms reveal the **shape of a distribution** — whether it is symmetric, skewed, or has multiple peaks. The number of bins controls the granularity.

A teacher recorded the following **test scores** for 20 students:


> 💡 **Hint:** Use `plt.subplot(1, 2, 1)` for the first and `plt.subplot(1, 2, 2)` for the second. Call `plt.tight_layout()` at the end.

## 📝 Task 7 — Interactive Line Plot with Bokeh

### Background
Bokeh creates **interactive, browser-rendered** plots. The `HoverTool` lets users see exact data values when they hover over points.

### 🎯 Your Task
Create an **interactive Bokeh line plot** showing website traffic over 7 days:

```
days    = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visitors = [120, 180, 150, 200, 250, 300, 270]
```

Requirements:
- Line width = 2, color of your choice
- Circle markers on each data point (size = 8)
- A `HoverTool` that shows the day and visitor count on hover
- Proper title and axis labels

> 💡 **Hint:** To use HoverTool with plain lists, convert data to a `ColumnDataSource`.  
> `source = ColumnDataSource(data=dict(days=days, visitors=visitors))`  
> Then use `p.line(x='days', y='visitors', source=source)` and  
> `HoverTool(tooltips=[("Day", "@days"), ("Visitors", "@visitors")])`.


````python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import HoverTool, ColumnDataSource

output_notebook()

days     = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visitors = [120, 180, 150, 200, 250, 300, 270]

# 1. Create ColumnDataSource
source = ColumnDataSource(data=dict(days=days, visitors=visitors))

# 2. Create figure
p = figure(title="Website Visitors Over 7 Days",
           x_axis_label="Days",
           y_axis_label="Number of Visitors",
           x_range=days)

# 3. Add line
p.line(x='days', y='visitors', source=source,
       line_width=2, color="blue")

# 4. Add circle markers
p.circle(x='days', y='visitors', source=source,
         size=8, color="red")

# 5. Add HoverTool
hover = HoverTool(tooltips=[("Day", "@days"),
                            ("Visitors", "@visitors")])
p.add_tools(hover)

# 6. Show plot
show(p)
````

````python
# YOUR CODE HERE — Part B: Bar chart for Asia in 2007
# Filter data
asia_2007 = gapminder[(gapminder["year"] == 2007) & (gapminder["continent"] == "Asia")]

# Sort values
asia_2007 = asia_2007.sort_values("lifeExp", ascending=False )

# Bar chart
fig = px.bar(
    asia_2007,
    x="country",
    y="lifeExp",
    color_discrete_sequence=['purple'],  # color set here

    title="Life Expectancy in Asia (2007)"
)

fig.show()
````

````python
# Q1: What is the main advantage of using Seaborn over raw Matplotlib for statistical plots?
# Seaborn provides high-level, aesthetically pleasing statistical plots with fewer lines of code. 
# It has built-in themes, color palettes, and functions for complex plots like violin, boxen, and pairplots, 
# making data visualization faster and more informative than raw Matplotlib.

# Q2: When would you choose Bokeh over Plotly, and vice versa?
# Bokeh is ideal when you want fully interactive, browser-based plots with Python integration, especially for large datasets 
# and server-side apps. Plotly is more versatile for quick interactive plots, supports animation and dashboards easily, 
# and integrates well with web apps like Dash. Choice depends on project complexity, web integration, and animation needs.

# Q3: In Task 3, how does changing the number of bins affect your interpretation of the distribution?
# Increasing bins gives more granularity and can show finer patterns, but too many bins may make the histogram noisy. 
# Fewer bins smooth out the distribution but may hide important variations. Choosing the right bin size balances clarity and detail.

# Q4: Why is a pie chart generally not recommended when you have more than 5–6 categories?
# With many categories, pie slices become too small and hard to distinguish. It is difficult to compare proportions accurately. 
# Bar charts or stacked charts are usually better for visual clarity.

# Q5: What does a violin plot show that a box plot does not?
# A violin plot shows the full probability density of the data at different values (distribution shape), 
# while a box plot only shows median, quartiles, and potential outliers. Violin plots reveal multimodal distributions and skewness.
````
