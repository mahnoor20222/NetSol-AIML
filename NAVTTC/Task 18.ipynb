# Basic continue in loop

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

#  Continue with Even Numbers

```python
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i)
```

# Continue with Odd Numbers

```python
salaries = [15000, 22000, 18000, 30000, 27000]

for salary in salaries:
    if salary % 2 != 0:
        continue
    print(salary)
```

# Continue with Real-Life Billing System

```python
bills = [120, 45, 300, 90, 55, 200]

for bill in bills:
    if bill < 100:
        continue
    print(bill)
```

# Continue with String (Real-Life Text Filtering)

```python
word = "PYTHON"

for letter in word:
    if letter == "O":
        continue
    print(letter)
```

#  Continue with List (Real-Life Attendance Filter)

```python
names = ["Ali", "Sara", "Ahmed", "Zara", "Usman"]

for name in names:
    if name.startswith("A"):
        continue
    print(name)
```

# Continue with Numbers (Discount Filter – Real Life)

```python
values = [100, 250, 80, 400, 50, 600]

for price in values:
    if price < 100:
        continue
    print(price)
```
