# Pandas DataFrames


```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df) 
```

# Locate Row


```python
#refer to the row index:
print(df.loc[0])
```

# Locate Named Indexes

```python
#refer to the named index:
print(df.loc["day2"])
```
