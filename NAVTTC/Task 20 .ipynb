# Function Arguments


```python
def evenOdd(x):
    if (x % 2 == 0):
        return "Even"
    else:
        return "Odd"

print(evenOdd(16))
print(evenOdd(7))
```

# Default Arguments

```python
def myFun(x,y= 50):
    print("x: ",x)
    print("y: ",y)
myFun(10)    
```

# Keyword Arguments

```python
def student(fname,lname):
    print(fname,lname)
student(fname='Geeks', lname='Practice')
student(lname='Practice', fname='Geeks')    
```

# Positional Arguments

```python
def nameAge(name, age):
    print("Hi, I am", name)
    print("My age is ", age)

print("Case-1:")
nameAge("Olivia", 27)

print("\nCase-2:")
nameAge(27, "Olivia")
```

# Arbitrary Arguments

```python
def myFun(*args, **kwargs):
    print("Non-Keyword Arguments (*args):")
    for arg in args:
        print(arg)

    print("\nKeyword Arguments (**kwargs):")
    for key, value in kwargs.items():
        print(f"{key} == {value}")

myFun('Hey', 'Welcome', first='Geeks', mid='for', last='Geeks')
```

# Function within Functions


```python
def f1():
    s = 'I love GeeksforGeeks'
    def f2():
        print(s)
        
    f2()
f1()
```
