```python
# Prints all letters except 'e' and 's'

for i in 'geeksforgeeks':

    if i == 'e' or i == 's':
        continue
    print(i)
```

```python
for i in 'geeksforgeeks':

    # break the loop as soon it sees 'e'
    # or 's'
    if i == 'e' or i == 's':
        break

print(i)
```

```python
# An empty loop
for i in 'geeksforgeeks':
    pass
print(i)
```

```python
for i in range(1, 4):
    print(i)
else:  # Executed because no break in for
    print("No Break\n")
```
