# Sorting Arrays


```python
import numpy as np

arr = np.array([3, 2, 0, 1])

print(np.sort(arr))
```

```python
import numpy as np

arr = np.array(['banana', 'cherry', 'apple'])

print(np.sort(arr))
```

```python
import numpy as np

arr = np.array([True, False, True])

print(np.sort(arr))
```

# Sorting a 2-D Array


```python
import numpy as np

arr = np.array([[3, 2, 4], [5, 0, 1]])

print(np.sort(arr))
```
