# House Price Prediction

```python
from sklearn.linear_model import LinearRegression

area = [[500], [800], [1000], [1200], [1500]]
price = [10000, 15000, 20000, 25000, 30000]

model = LinearRegression()
model.fit(area, price)

prediction = model.predict([[1100]])
print(prediction)
```

# Temperature vs Sales

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

temp = [[20], [25], [30], [35], [40]]
sales = [200, 300, 400, 500, 600]

model = LinearRegression()
model.fit(temp, sales)

prediction = model.predict([[32]])
print(prediction)

plt.scatter(temp, sales)      # actual data points
plt.plot(temp, model.predict(temp))  # regression line
plt.xlabel("Temperature")
plt.ylabel("Sales")
plt.show()
```

# Model Evaluation

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

X = [[1], [2], [3], [4], [5]]
y = [20, 40, 50, 70, 90]

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("R2 Score:", r2)
```
