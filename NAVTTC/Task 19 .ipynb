# Using enumerate()


```python
words=['The','Big','Bag','Thory']
for index,values in enumerate (words):
  print(index,values)
```

# Using zip()


```python
p1 = {'Washington': 'First', 'Lincoln': 'Emancipator', 'Roosevelt': 'Trust-Buster'}

for name, title in p1.items():
    print(name, title)
```

# Using sorted()


```python
num = [1, 3, 5, 6, 2, 1, 3]

print("Sorted list:")
for n in sorted(num):
     print(n, end=" ")


print("\nSorted list without duplicates:")
for n in sorted(set(num)):
    print(n, end=" ")
```

# Using reversed()


```python
num = [1, 3, 5, 6, 2, 1, 3]
for n in reversed(num):
    print(n, end= '')
```

# While Loop with If Condition


```python
count = 0

while count < 5:
    if count == 3:
        print("Count is 3")
    count += 1
```
