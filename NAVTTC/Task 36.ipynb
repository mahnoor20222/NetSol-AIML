# Exercise on Numpy Array iteration:

```python
import numpy as np
arr = np.array([1,2,3])

for x in arr:
    print(x)
```

```python
arr = np.array([[1,2,3],[4,5,6]])

for x in arr:
    print(x)
```

```python
for x in arr:
    for y in x:
        print(y)
```

```python
for x in np.nditer(arr):
    print(x)
```

```python
for index, value in np.ndenumerate(arr):
    print(index, value)
```
